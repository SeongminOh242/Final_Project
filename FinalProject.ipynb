#### **I calculated the employment rate(%): (the number of employed people age of 16+) / Total Population.**
#### **Since the data is big, I only calculate top 20 of them.**
#### **In summary, the script provides a concise analysis and visualization of employment rates in the top 20 neighborhoods, offering insights into the distribution of employment across these areas.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
file = pd.read_csv("employment.csv")
file = file.head(20)

# Clean and convert columns to appropriate data types
file['Population'] = file['Population (2010)'].str.replace(',', '').astype(int)
file['Employment'] = file['Total Age 16+ N\'hood Residents Employed (2010 Est.)'].str.replace(',', '').astype(int)

# Calculate Employment Rate and create a new column
file['Employment Rate'] = (file['Employment'] / file['Population']) * 100

# Sort the DataFrame based on the 'Employment Rate' column
file_sorted = file.sort_values(by='Employment Rate', ascending=True)

# Plotting the bar graph with color customization
plt.figure(figsize=(12, 8))
bars = plt.bar(file_sorted['Neighborhood'], file_sorted['Employment Rate'], color='skyblue')

# Adding color gradient to the bars
for bar in bars:
    bar.set_color(plt.cm.viridis(bar.get_height() / 100))

plt.title('Top 20 Neighborhoods by Employment Rate')
plt.xlabel('Neighborhood')
plt.ylabel('Employment Rate')
plt.xticks(rotation='vertical')
plt.show()
```
